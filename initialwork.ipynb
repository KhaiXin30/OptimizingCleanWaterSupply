```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('water_quality_training_dataset.csv')
df.head()
```

```python
df_numeric = df.select_dtypes(include=[np.number])
df_numeric.corr()
```

```python
sns.heatmap(df_numeric.corr(), annot=True, cmap='coolwarm')
```

```python
df_time = df[(df["Latitude"] == -28.760833) & (df["Longitude"] == 17.730278)]
```

```python
plt.plot(df_time['Sample Date'], df_time['Total Alkalinity'], label='Total Alkalinity')
plt.plot(df_time['Sample Date'], df_time['Electrical Conductance'], label='Electrical Conductance')
plt.plot(df_time['Sample Date'], df_time['Dissolved Reactive Phosphorus'], label='Dissolved Reactive Phosphorus')
plt.xlabel('Sample Date')
plt.ylabel('Values')
plt.title('Total Alkalinity, Electrical Conductance, and Dissolved Reactive Phosphorus over Time at Specific Location')
plt.legend()
plt.xticks(rotation=45)
plt.show()
```
